```python
def crossVelPlot(d00, d16, d30, d50, d70, n00, n16, n30, n50, n70):
    plt.plot(n00.arc_length-4.0, n00.UMean_X, 'r--', linewidth=0.95)
    plt.plot(d00.arc_length-4.0, d00.UMean_X, 'b-', linewidth=0.95)
    plt.plot(n16.arc_length-4.0, n16.UMean_X + .5, 'r--', linewidth=0.95)
    plt.plot(d16.arc_length-4.0, d16.UMean_X + .5, 'b-', linewidth=0.95)
    plt.plot(n30.arc_length-4.0, n30.UMean_X + .85, 'r--', linewidth=0.95)
    plt.plot(d30.arc_length-4.0, d30.UMean_X + .85, 'b-', linewidth=0.95)
    plt.plot(n50.arc_length-4.0, n50.UMean_X + 1, 'r--', linewidth=0.95)
    plt.plot(d50.arc_length-4.0, d50.UMean_X + 1, 'b-', linewidth=0.95)
    plt.plot(n70.arc_length-4.0, n70.UMean_X + 1.15, 'r--', linewidth=0.95)
    plt.plot(d70.arc_length-4.0, d70.UMean_X + 1.15, 'b-', linewidth=0.95)

    plt.legend(['NN', 'CFD'])
    plt.xlabel(r'$y/D$')
    plt.ylabel(r'$U_x$')
    plt.tick_params(axis='y', which='both',
                    left=False, right=False, labelleft=False)
    plt.text(2.55, 0.92, r'$x/D = 0.0$')
    plt.text(2.55, 1.43, r'$x/D = 1.6$')
    plt.text(2.55, 1.77, r'$x/D = 3.0$')
    plt.text(2.55, 1.92, r'$x/D = 5.0$')
    plt.text(2.55, 2.07, r'$x/D = 7.0$')
```
