```python
def energyCal(var):
    n = len(var)
    var_hat = fft(var - np.mean(var))
    var_hat = var_hat[0:int(n/2+1)]
    s = 2*np.pi/n
    f = np.arange(0, np.pi, s)
    e = 1/(2*np.pi*n) * np.abs(var_hat)**2.0
     
    return f, e

def engPlot(nnVar, dynVar, title, ylab):
    nnf, nne = energyCal(nnVar)
    dynf, dyne = energyCal(dynVar)
    plt.loglog(nnf,10**(-4.3) * nnf**(-3), 'k--')
    plt.loglog(dynf,dyne)
    plt.loglog(nnf,nne)
    plt.ylim([1e-10, 1e3])
    plt.legend(['$k^{-5/3}$', 'CFD', 'NN'])
    plt.title(title)
    plt.xlabel(r'freq')
    plt.ylabel(ylab)
```

```python
engPlot(nnC1.U_X.values, dynC1.U_X.values, r'$\mathcal{R}e=1 \times 10^3$', r'$U_x$')
```
